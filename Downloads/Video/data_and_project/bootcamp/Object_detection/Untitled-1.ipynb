```python
import keras 
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.models import Model
from keras.applications.vgg16 import VGG16
from keras.applications.vgg16 import preprocess_input
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
import tensorflow as tf

import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
train_data='train'
test_data='test'
```

```python
# use image data Generator to import the images from the dataset
train_gen=ImageDataGenerator(rescale=1./255,
                                 shear_range = 0.2,
                                 zoom_range = 0.2,
                                 horizontal_flip = True)

test_gen = ImageDataGenerator(rescale = 1./255)
train_batches = train_gen.flow_from_directory(   train_data,
                                                 target_size=(256,256),
                                                 class_mode='sparse',
                                                 batch_size=4,
                                                 shuffle=True,
                                                 color_mode = 'rgb')

test_batches = test_gen.flow_from_directory(    test_data,
                                                target_size=(256,256),
                                                 class_mode='sparse',
                                                 batch_size=4,
                                                 shuffle=False,
                                                 color_mode = 'rgb')
```

```python
train_batch=train_batches[0]
print(train_batch[0].shape)
print(train_batch[1])

test_batch=test_batches[0]
print(test_batch[0].shape)
print(test_batch[1])
```

```python
import os
import glob

# Set the directory path where your training image data is stored
train_data_dir = 'train'

# Get all subdirectories (i.e., class labels) within the training data directory
subdirs = [d for d in os.listdir(train_data_dir) if os.path.isdir(os.path.join(train_data_dir, d))]

# Initialize lists to store image file paths and corresponding labels
image_file_paths = []
names = []

# Loop through each subdirectory (class label)
for subdir in subdirs:
    # Get all image file paths within the subdirectory
    image_files = glob.glob(os.path.join(train_data_dir, subdir, "*.jpg")) # change "*.jpg" to the appropriate file extension of your images
    # Extract the label from the subdirectory name
    label = subdir
    # Append image file paths and labels to the respective lists
    image_file_paths.extend(image_files)
    names.extend([label] * len(image_files))

```

```python
def show_images(batch, pred_labels=None, names=None):
    plt.figure(figsize=(10, 10))

    for i in range(4):
        plt.subplot(2, 2, i+1)  # Fixed to subplot() function call
        plt.xticks([])
        plt.yticks([])
        plt.grid(False)
        plt.imshow(batch[0][i], cmap=plt.cm.binary)
        lbl = names[int(batch[1][i])]  # Removed unnecessary int() function call
        
        if pred_labels is not None and names is not None:
            true_lbl = int(batch[1][i])
            pred_lbl = int(pred_labels[i])
            lbl += f'/pred:{names[pred_lbl]}'
            
        plt.xlabel(lbl)

    plt.show()

pred_labels = [0, 1, 2, 3]  
names = ['label0', 'label1', 'label2', 'label3']
# Call the function with a test batch of images
show_images(test_batch, pred_labels, names)
```

```python
# Define the desired size of the input images
IMAGE_SIZE = (256, 256)

# add preprocessing layer to the front of VGG
vgg = VGG19(input_shape=IMAGE_SIZE + (3,), weights='imagenet', include_top=False)

# don't train existing weights
for layer in vgg.layers:
    layer.trainable = False
    
# useful for getting number of classes
folders = glob('train')

# our layers
x = Flatten()(vgg.output)
prediction = Dense(len(folders), activation='softmax')(x)

# create a model object
model = Model(inputs=vgg.input, outputs=prediction)

# view the structure of the model
model.summary()
```

```python
# compile the model

model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics='accuracy'
)
```

```python
# Saving the trained model 
model.save('Animal_detection.h5')
```

```python
# Evaluating the model on test batches
model.evaluate(test_batches,verbose=2)
```

```python
# Making predictions using the trained model
predictions = model.predict(test_batches)

# Applying softmax activation to the predictions
predictions = tf.nn.softmax(predictions)

# Extracting labels from the predictions
labels = np.argmax(predictions, axis=1)
```

```python
# Displaying images with their corresponding labels
show_images(test_batches[0], labels[0:4])
```
